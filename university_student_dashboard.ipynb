```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = '/content/university_student_dashboard_data.csv'
data = pd.read_csv(url)

print(data.head())
print(data.info())

data['Year-Term'] = data['Year'].astype(str) + ' ' + data['Term']
data['Admission Rate (%)'] = (data['Admitted'] / data['Applications']) * 100
data['Enrollment Rate (%)'] = (data['Enrolled'] / data['Admitted']) * 100
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(x='Year-Term', y='Retention Rate (%)', data=data, marker='o', color='purple')
plt.xticks(rotation=45)
plt.title('Retention Rate Trends Over Time')
plt.ylabel('Retention Rate (%)')
plt.xlabel('Year-Term')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Year-Term', y='Student Satisfaction (%)', data=data, palette='Greens')
plt.xticks(rotation=45)
plt.title('Student Satisfaction Over the Years')
plt.xlabel('Year-Term')
plt.ylabel('Satisfaction (%)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
data_melted = data.melt(id_vars='Year-Term', value_vars=['Engineering Enrolled', 'Business Enrolled', 'Arts Enrolled', 'Science Enrolled'],
                        var_name='Department', value_name='Enrollment')
sns.lineplot(x='Year-Term', y='Enrollment', hue='Department', data=data_melted, marker='o') # Changed y to 'Enrollment'
plt.xticks(rotation=45)
plt.title('Enrollment Trends by Department')
plt.xlabel('Year-Term')
plt.ylabel('Enrollment Numbers')
plt.legend(title='Department')
plt.show()
```

```python
spring_data = data[data['Term'] == 'Spring']
fall_data = data[data['Term'] == 'Fall']

plt.figure(figsize=(14, 7))
plt.plot(spring_data['Year'], spring_data['Enrolled'], label='Spring Enrollment', marker='o')
plt.plot(fall_data['Year'], fall_data['Enrolled'], label='Fall Enrollment', marker='s')
plt.title('Spring vs. Fall Enrollment Trends')
plt.xlabel('Year')
plt.ylabel('Enrolled Students')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Retention Rate (%)', y='Student Satisfaction (%)', hue='Term', data=data, s=120)
plt.title('Correlation Between Retention Rate and Student Satisfaction')
plt.xlabel('Retention Rate (%)')
plt.ylabel('Student Satisfaction (%)')
plt.grid(True)
plt.legend(title='Term')
plt.show()
```



### **What the Data is Telling Us About Students & Universities**

1. **How Well Are Students Sticking Around?** *(Retention Rate Trends Over Time)*  
   - This graph shows whether students are staying enrolled after their first year. If the retention rate is climbing, that’s great news—it means students are feeling supported and engaged.  
   - But if it’s dropping, there could be issues like tougher coursework, financial struggles, or even dissatisfaction with campus life. A declining trend might be a red flag for university administrators to step in and make improvements.

2. **Are Students Actually Happy?** *(Student Satisfaction Over the Years)*  
   - This bar chart tracks how students feel about their university experience over time.  
   - If satisfaction is rising, the university might be doing a better job with facilities, academics, or student life.  
   - But if it’s declining, students could be feeling overwhelmed, unsupported, or dissatisfied with their classes or services. It’s a crucial metric because happy students are more likely to succeed and stay enrolled.

3. **Which Departments Are Attracting Students?** *(Enrollment Trends by Department)*  
   - This chart breaks down how many students are enrolling in different majors like Engineering, Business, Arts, and Science.  
   - If one department is booming while others struggle, it could mean job market trends are influencing students’ choices.  
   - A drop in enrollment in certain fields might indicate that students see fewer opportunities in those areas, pushing universities to rethink their curriculum.

4. **When Are Students Choosing to Start?** *(Spring vs. Fall Enrollment Trends)*  
   - Traditionally, Fall is the big intake season, but more students starting in Spring could mean the university is offering more flexible admission options.  
   - If there’s a big shift in these patterns, it might suggest students are choosing alternative pathways, like gap years, work experience, or online courses before enrolling full-time.



**Final Thoughts:**  
These insights help universities understand what’s working and what’s not. If students are dropping out, they need better support systems. If certain majors are declining, they need to revamp programs to match industry trends. And if student satisfaction is dipping, well—that’s a sign to listen to student feedback and make real changes.
